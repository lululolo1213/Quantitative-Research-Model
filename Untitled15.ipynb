# Portfolio Construction

```python
# Portfolio selection
N = 20  # number of stocks to select
portfolio_df = pd.DataFrame()

# Convert 'fd' to datetime and extract the month in filtered_df
filtered_df['fd'] = pd.to_datetime(filtered_df['fd'].astype(str), format='%Y-%m-%d')
filtered_df['month'] = filtered_df['fd'].dt.to_period('M')

# For each month, select top N stocks based on the rank and add to the portfolio
for month in filtered_df['month'].unique():
    monthly_stocks = filtered_df[filtered_df['month'] == month].nlargest(N, 'rank')
    portfolio_df = pd.concat([portfolio_df, monthly_stocks])

# Print portfolio details
print(portfolio_df)
```

```python
# Convert 'td' in market_df to datetime format and extract the month
market_df['td'] = pd.to_datetime(market_df['td'].astype(str), format='%Y-%m-%d')
market_df['month'] = market_df['td'].dt.to_period('M')

# Extract month from 'fd' in portfolio_df
portfolio_df['month'] = portfolio_df['fd'].dt.to_period('M')

# Merge portfolio_df with market_df to get the close prices of the selected stocks
portfolio_df = pd.merge(portfolio_df, market_df[['td', 'codenum', 'close', 'month']], 
                        left_on=['codenum', 'month'], 
                        right_on=['codenum', 'month'])

# Get last day's close price of each month for each stock
last_day_close = market_df.sort_values('td').groupby(['codenum', 'month'])['close'].last()
 
# Calculate monthly returns of all stocks
all_returns = last_day_close.groupby('codenum').pct_change()

# Calculate monthly returns of selected stocks
portfolio_df['monthly_return'] = portfolio_df.groupby('codenum')['close'].pct_change()

# Calculate equally weighted portfolio return
N = len(portfolio_df['codenum'].unique())
portfolio_df['weight'] = 1/N
portfolio_df['weighted_return'] = portfolio_df['weight'] * portfolio_df['monthly_return']
portfolio_return = portfolio_df.groupby('month')['weighted_return'].sum()
average_return = all_returns.groupby('month').mean()

# Compare portfolio return with the average return in the pool
comparison_df = pd.concat([portfolio_return, average_return], axis=1)
comparison_df.columns = ['portfolio_return', 'average_return']

print(comparison_df)
```

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

comparison_df = comparison_df[comparison_df.index.year >= 2000]


fig, ax = plt.subplots(figsize=(10, 6))

comparison_df[['portfolio_return', 'average_return']].plot(ax=ax)
ax.yaxis.set_major_formatter(FuncFormatter(lambda y, _: '{:.0%}'.format(y))) 


cumulative_return = (1 + comparison_df['portfolio_return']).cumprod() - 1
cumulative_return.plot(ax=ax)
plt.fill_between(comparison_df.index, cumulative_return, 0, 
                 where=cumulative_return<0, 
                 color='black', alpha=0.3)
max_dd = (cumulative_return.div(cumulative_return.cummax())-1).min()
plt.title('Returns and Maximum Drawdown\nMaximum Drawdown: {:.2%}'.format(max_dd))

plt.show()
```
